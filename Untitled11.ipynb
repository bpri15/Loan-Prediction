```python
combined_data = pd.concat([df_train,df_test], axis=0)
combined_data
```

```python
combined_data['Gender'].fillna('Male',inplace=True)
combined_data['Married'].fillna('Yes', inplace=True)
combined_data['Dependents'].fillna(0, inplace=True)
combined_data['Self_Employed'].fillna('No', inplace=True)
combined_data['LoanAmount'].fillna(df_train['LoanAmount'].mean(),inplace=True)
combined_data['Loan_Amount_Term'].fillna(df_train['Loan_Amount_Term'].mean(), inplace=True)
combined_data['Credit_History'].fillna(1.0, inplace=True)
combined_data.isnull().sum()
```

```python
from sklearn import preprocessing
# label_encoder object knows how to understand word labels. 
label_encoder = preprocessing.LabelEncoder()
# Encode labels in column 'Country'. 
combined_data['Gender'] = label_encoder.fit_transform(combined_data['Gender']) 
combined_data['Married']= label_encoder.fit_transform(combined_data['Married']) 
combined_data['Education']= label_encoder.fit_transform(combined_data['Education']) 
combined_data['Self_Employed']= label_encoder.fit_transform(combined_data['Self_Employed']) 
combined_data['Property_Area']= label_encoder.fit_transform(combined_data['Property_Area']) 
combined_data
```

```python
combined_data.replace('3+',3, inplace=True)
```

```python
x_train = combined_data.iloc[:476,1:-1]
x_test = combined_data.iloc[476:,1:-1]
y_train = combined_data.iloc[:476,-1]
y_test = combined_data.iloc[476:,-1]
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression  
classifier= LogisticRegression(random_state=0)  
classifier.fit(x_train, y_train)  
```

```python
y_pred= classifier.predict(x_test)  
y_pred
```

```python
df_test['Loan_Status'] = y_pred
```

```python
df_test.to_csv('/content/generated_Final.csv', columns=['Loan_ID', 'Loan_Status'])
```
